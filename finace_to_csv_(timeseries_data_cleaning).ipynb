```python
# spliting amazon's data into train and testing data
split_size_AMZN = int(0.8 * len(amazon_price)) 

# Create train data splits (everything before the split)
X_train_AMZN, y_train_AMZN = timesteps[:split_size_AMZN], amazon_price[:split_size_AMZN]

# Create test data splits (everything after the split)
X_test_AMZN, y_test_AMZN = timesteps[split_size_AMZN:], amazon_price[split_size_AMZN:]

len(X_train_AMZN), len(X_test_AMZN), len(y_train_AMZN), len(y_test_AMZN)
```

```python
def plot_time_series(timesteps, values, format='.', start=0, end=None, label=None):
  plt.plot(timesteps[start:end], values[start:end], format, label=label)
  plt.xlabel("Time")
  plt.ylabel("BTC Price")
  if label:
    plt.legend(fontsize=14) # make label bigger
  plt.grid(True)
```

```python

plt.figure(figsize=(10, 7))
plot_time_series(timesteps=X_train_AMZN, values=y_train_AMZN, label="Train data")
plot_time_series(timesteps=X_test_AMZN, values=y_test_AMZN, label="Test data")
```
